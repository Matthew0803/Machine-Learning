```python
import pandas as pd
datas = pd.read_csv('celcius.csv')
```

```python
datas.info()
```

```python
datas.head()
```

```python
import seaborn as sns
sns.scatterplot(x='celsius',y=' farenheit',data=datas,hue=' farenheit',palette='coolwarm')
```

```python
#characteristics (X), label(y)
x =datas['celsius']
y =datas[' farenheit']
```

```python
print(x)
```

```python
type(x)
```

```python
x.values
```

```python
xProcessed = x.values.reshape(-1,1)
yProcessed = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(xProcessed,yProcessed)
#training
```

```python
celsius = 900
prediction = model.predict([[celsius]])
print(prediction)

#prediction
```

```python
model.score(xProcessed,yProcessed)
#score, 1.0 means perfect
```
