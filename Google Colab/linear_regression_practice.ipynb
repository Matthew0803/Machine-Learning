```python
import pandas as pd
datas = pd.read_csv('random_height_weight.csv')
datas.head()
#print data in form
```

```python
datas.info()
#make sure data is fine
```

```python
import seaborn as sns
model = sns.scatterplot(x = 'height_cm', y = 'weight_kg', data = datas,
    hue='weight_kg', palette='coolwarm')
```

```python
x = datas['height_cm']
xProcess = x.values.reshape(-1,1)
y = datas['weight_kg']
yProcess = y.values.reshape(-1,1)
#reshape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(xProcess,yProcess)
#training
```

```python
#predict
prediction = 150
model.predict([[prediction]])
```

```python
#score
model.score(xProcess,yProcess)
```
